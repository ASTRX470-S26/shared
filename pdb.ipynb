# Debugging with `pdb` in Python

`pdb` is Python's built-in interactive debugger. It lets you pause execution, step through code, inspect variables, and find bugs.

## Key commands

| Command | Description |
|--------|-------------|
| `n` (next) | Execute the next line |
| `s` (step) | Step into a function call |
| `c` (continue) | Continue until next breakpoint |
| `p <expr>` | Print the value of an expression |
| `l` (list) | Show source code around current line |
| `q` (quit) | Quit the debugger |
| `b <line>` | Set a breakpoint at a line number |
| `b <line> if <cond>` | Set a conditional breakpoint |

## Example 2: Debugging in a loop

Here we have a function that calculates the average of a list, but it has a bug — it divides by the wrong value. We'll use `pdb` to inspect what's happening inside the loop.

```python
import pdb

def calculate_average(numbers):
    total = 0
    for i, n in enumerate(numbers):
        total += n
        pdb.set_trace()  # pause on every iteration to inspect
    average = total / i  # Bug: should be len(numbers), not i
    return average

data = [10, 20, 30, 40, 50]
result = calculate_average(data)
print(f"Average: {result}")  # should be 30.0, but will be wrong
```

While paused, try these commands:
- `p i` — print current index
- `p total` — print running total
- `p n` — print current number
- `c` — continue to next iteration
- `q` — quit

## Example 3: Post-mortem debugging

`pdb.post_mortem()` lets you inspect the state of the program *after* an exception has occurred, without needing to set breakpoints in advance.

```python
import pdb, traceback, sys

def divide(a, b):
    return a / b

def process():
    values = [10, 5, 0, 2]
    results = []
    for v in values:
        results.append(divide(100, v))  # will crash when v = 0
    return results

try:
    process()
except Exception:
    pdb.post_mortem()  # drop into debugger at the point of the crash
```

While in post-mortem mode:
- `l` — see the line that caused the crash
- `p a, b` — inspect the values that caused the `ZeroDivisionError`
- `u` (up) — move up the call stack to `process()`
- `p v` — see which value in the list triggered it
- `q` — quit
